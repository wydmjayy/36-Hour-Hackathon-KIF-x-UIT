```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/jaypatel/Desktop/hackathon/36-Hour-Hackathon-KIF-x-UIT/subscription_surgeon_dataset.csv")
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
# Total spending
total_spending = df["amount"].sum()
print("Total Spending:", total_spending)
```

```python
# Spending by category
category_spending = df.groupby("category")["amount"].sum()
print("\nCategory Spending:\n", category_spending)
```

```python
# Spending by merchant
merchant_spending = df.groupby("merchant")["amount"].sum().sort_values(ascending=False)
print("\nTop Subscriptions:\n", merchant_spending.head())
```

```python
# Frequency (how many times paid)
frequency = df["merchant"].value_counts()
```

```python
# Household duplication
duplicates = df.groupby(["household_id", "merchant"]).size()
duplicates = duplicates[duplicates > 1]

print("\nDuplicates:\n", duplicates)
```
